### TPC3: Jogo dos 21 fósforos

```python
import random

print("Bem-vindo ao jogo dos 21 fósforos!")

modo=int(input("Escolha o modo que pretende jogar: 1- o JOGADOR joga em primeiro lugar; 2- o COMPUTADOR joga em primeiro lugar"))

while modo!=1 and modo!=2:
    print("\nResposta inválida! Introduza o modo 1 ou 2.")
    modo=int(input("Escolha o modo que pretende jogar: 1- o JOGADOR joga em primeiro lugar; 2- o COMPUTADOR joga em primeiro lugar"))

if modo==1:
    fosforos=21
    print("Há 21 fósforos!")
    print("É você a começar.")

    while fosforos>0:
        retirar=int(input("Deseja tirar 1, 2, 3 ou 4 fósforos?"))

        while retirar>4 or retirar<1 or retirar>fosforos:
            print("\nResposta inválida! Apenas pode retirar 1, 2, 3 ou 4 fósforos e não pode tirar mais fósforos que os restantes.")
            retirar=int(input("Deseja tirar 1, 2, 3 ou 4 fósforos?"))
        
        print(f"\nRemoveu {retirar} fósforos.")

        fosforos=fosforos-retirar
        print(f"Restam {fosforos} fósforos!")


        if fosforos>5:
            vezcomputador=5-retirar
            fosforos=fosforos-vezcomputador
            print(f"\nO computador removeu {vezcomputador} fósforos.")
            print(f"Restam {fosforos} fósforos!")
        elif fosforos>0:
            vezcomputador=fosforos-1
            fosforos=fosforos-vezcomputador
            print(f"\nO computador removeu {vezcomputador} fósforos.")
            print(f"Restam {fosforos} fósforos!")

        if fosforos==0:
            print("\nRetirou o último fósforo. Perdeu o jogo!")



if modo==2:
    fosforos=21
    print("É o computador a começar. Há 21 fósforos!")

    vezcomputador=random.randint(1,4)
    print(f"\nO computador removeu {vezcomputador} fósforos.")
    fosforos=fosforos-vezcomputador
    print(f"Restam {fosforos} fósforos!")
    
    retirar=int(input("Deseja tirar 1, 2, 3 ou 4 fósforos?"))

    while retirar>4 or retirar<1 or retirar>fosforos:
            print("\nResposta inválida! Apenas pode retirar 1, 2, 3 ou 4 fósforos e não pode tirar mais fósforos que os restantes.")
            retirar=int(input("Deseja tirar 1, 2, 3 ou 4 fósforos?"))
        
    print(f"\nRemoveu {retirar} fósforos.")

    fosforos=fosforos-retirar
    print(f"Restam {fosforos} fósforos!")
        
    while fosforos==6 or fosforos==11 or fosforos==16:
        vezcomputador=random.randint(1,4)
        print(f"\nO computador removeu {vezcomputador} fósforos.")
        fosforos=fosforos-vezcomputador
        print(f"Restam {fosforos} fósforos!")

        retirar=int(input("Deseja tirar 1, 2, 3 ou 4 fósforos?"))
        while retirar>4 or retirar<1 or retirar>fosforos:
            print("\nResposta inválida! Apenas pode retirar 1, 2, 3 ou 4 fósforos e não pode tirar mais fósforos que os restantes.")
            retirar=int(input("Deseja tirar 1, 2, 3 ou 4 fósforos?"))

        print(f"\nRemoveu {retirar} fósforos.")
        fosforos=fosforos-retirar
        print(f"Restam {fosforos} fósforos!")

        if fosforos==1:
            print("\nO computador retirou o último fósforo. O computador perdeu o jogo!")

    if fosforos!=6 and fosforos!=11 and fosforos!=16 and fosforos>1:
        vezcomputador=0
        while fosforos>1:
            while fosforos!=6 and fosforos!=11 and fosforos!=16:
                vezcomputador=vezcomputador+1
                fosforos=fosforos-1
            print(f"\nO computador removeu {vezcomputador} fósforos.")
            print(f"Restam {fosforos} fósforos!")
            retirar=int(input("Deseja tirar 1, 2, 3 ou 4 fósforos?"))
            while retirar>4 or retirar<1 or retirar>fosforos:
                print("\nResposta inválida! Apenas pode retirar 1, 2, 3 ou 4 fósforos e não pode tirar mais fósforos que os restantes.")
                retirar=int(input("Deseja tirar 1, 2, 3 ou 4 fósforos?"))
            print(f"\nRemoveu {retirar} fósforos.")
            fosforos=fosforos-retirar
            print(f"Restam {fosforos} fósforos!")
            vezcomputador=5-retirar
            fosforos=fosforos-vezcomputador
        if fosforos==1:
                print(f"\nO computador removeu {vezcomputador} fósforos.")
                print(f"Restam {fosforos} fósforos!")
                print("\nTeve de retirar o último fósforo. Perdeu o jogo!")
```
